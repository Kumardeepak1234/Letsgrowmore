# Task-4 Image to Pencil Sketch with Python

```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Load the image
image_path = "images.jpeg"
original_img = cv2.imread(image_path)
```

```python
cv2_imshow(original_img)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Invert the grayscale image to create a negative effect
negative_img = cv2.bitwise_not(gray_img)
```

```python
cv2_imshow(negative_img)
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur to the negative image
blurred_img = cv2.GaussianBlur(negative_img, (21, 21), sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(blurred_img)
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```

```python
# Divide the grayscale image by the blurred negative image to create the sketch
sketch_img = cv2.divide(gray_img, 255 - blurred_img, scale=256)
```

```python
cv2_imshow(sketch_img)
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```
